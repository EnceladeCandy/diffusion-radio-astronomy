```python
import matplotlib.pyplot as plt 
import torch
plt.style.use("dark_background")

num_rows = 5
num_cols = 5
fig, axs = plt.subplots(5, 5, figsize = (8, 8))
x = torch.load("../../prior.pt", map_location = "cpu")
norm_x = (x + 1)/2

k = 0
for i in range(num_rows):
    for j in range(num_cols):
        axs[i, j].imshow(norm_x[k, ...].cpu(), "hot")
        axs[i, j].axis("off")
        axs[i, j].annotate(f"{k}", (0.05, 0.8), xycoords = "axes fraction", color = "white", fontsize = 10)
        k+=1
```
